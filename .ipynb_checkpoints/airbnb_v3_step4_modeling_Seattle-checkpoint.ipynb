```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option("display.max.columns", None)
pd.set_option('min_rows', 300)
# pd.set_option('display.precision',4)
```

```python
df = pd.read_csv('results\df_listings_amenity_fillna_dummy.csv', sep=',', index_col='id')
df.head(5)
```

```python
y = df[df.city_Seattle == 1]['price']
X = df[df.city_Seattle == 1].drop(columns=['price', 'city_Seattle'], axis=1)
```

```python
#Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40) 

lm_model = LinearRegression(normalize=True) # Instantiate
lm_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = lm_model.predict(X_train)
y_test_preds = lm_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
lm_model.coef_
```

```python
ridge_model = Ridge(alpha=0.01, normalize=True) # Instantiate
ridge_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = ridge_model.predict(X_train)
y_test_preds = ridge_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
ridge_model = Ridge(alpha=1, normalize=True) # Instantiate
ridge_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = ridge_model.predict(X_train)
y_test_preds = ridge_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
lasso_model = Lasso(alpha=0.15, normalize=True, max_iter=10000) # Instantiate
lasso_model.fit(X_train, y_train) #Fit
        
train_score=lasso_model.score(X_train,y_train)
test_score=lasso_model.score(X_test,y_test)
coeff_used = np.sum(lasso_model.coef_!=0)

print("training score: {}".format(train_score))
print("test score: {}".format(test_score))
print("number of features used: {}".format(coeff_used))
```

```python
lasso_model.coef_
```

```python
###Model Lasso regression
coef = pd.Series(lasso_model.coef_, index = X_train.columns)
coef[coef!=0].sort_values(ascending=False)
```

```python
imp_coef = pd.concat([coef.sort_values().head(3),
                     coef.sort_values().tail(5)])
plt.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Coefficients in the Lasso Model")
```
