```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

from tqdm import tqdm

import seaborn as sns
%matplotlib inline

from geopy.distance import geodesic
```

```python
def calculate_distance(df):
    boston_center_lat_lon = (42.3584300, -71.0597700)
    seattle_center_lat_lon = (47.6062100, -122.3320700)
    rental_loc = (df.latitude, df.longitude)
    
    if df.city == 'Boston':
        distance = geodesic(boston_center_lat_lon, rental_loc).miles
    elif df.city == 'Seattle':
        distance = geodesic(seattle_center_lat_lon, rental_loc).miles
    else:
        distance = 'is the city correct?'
    return distance
```

```python
df['distance_to_city_center'] = df.apply(lambda loc: calculate_distance(loc), axis=1)
```

```python
df['host_response_time'].fillna(df['host_response_time'].mode().iloc[0], inplace=True)
```

```python
df.host_response_time.value_counts()
```

```python
y = df['price']
```

```python
X = df.drop(columns=['last_scraped','host_id','host_since','zipcode','latitude', 'longitude', 'price'], axis=1)
```

```python
X['amenities'].replace(to_replace='[\{\}\"]', value='', regex=True, inplace=True)
amenities_dummy = X['amenities'].str.get_dummies(sep=',').add_prefix('amenities_')
```

```python
X.drop(columns=['amenities'], inplace=True, axis=1)
```

```python
cols_cat = X.select_dtypes(include=['object']).columns.to_list()
```

```python
cols_cat
```

```python
X_dummy = pd.get_dummies(data=X, prefix=cols_cat, prefix_sep='_', columns=cols_cat, drop_first=True)
```

```python
X = pd.concat([X_dummy, amenities_dummy], axis=1, sort=False)
X.head()
# X.shape
```

```python
X.isnull().sum().sort_values()
```

```python
#Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40) 

lm_model = LinearRegression(normalize=True) # Instantiate
lm_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = lm_model.predict(X_train)
y_test_preds = lm_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
ridge_model = Ridge(alpha=0.001, normalize=True) # Instantiate
ridge_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = ridge_model.predict(X_train)
y_test_preds = ridge_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
lasso_model = Lasso(alpha=0.09, normalize=True, max_iter=10000) # Instantiate
lasso_model.fit(X_train, y_train) #Fit
        
train_score=lasso_model.score(X_train,y_train)
test_score=lasso_model.score(X_test,y_test)
coeff_used = np.sum(lasso_model.coef_!=0)

print("training score: {}".format(train_score))
print("test score: {}".format(test_score))
print("number of features used: {}".format(coeff_used))
```
