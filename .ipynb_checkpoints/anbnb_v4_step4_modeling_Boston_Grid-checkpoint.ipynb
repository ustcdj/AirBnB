```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
from sklearn.model_selection import KFold, StratifiedKFold, ShuffleSplit, cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option("display.max.columns", None)
pd.set_option('min_rows', 300)
# pd.set_option('display.precision',4)
```

```python
df = pd.read_csv('results\df_listings_boston.csv', sep=',', index_col='id')
df.head(5)
```

```python
y = df['price']
X = df.drop(columns=['price'], axis=1)
```

```python
#Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0) 

lm_model = LinearRegression() # Instantiate
lm_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = lm_model.predict(X_train)
y_test_preds = lm_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
plt.scatter(y_train, y_train_preds)
```

```python
coef = pd.Series(lm_model.coef_, index = X_train.columns)
coef.sort_values(ascending=False)
```

```python
# use RidgeCV to find best alpha automatically
from sklearn.linear_model import RidgeCV

ridge_model = RidgeCV(alphas=(0.01, 0.05, 0.1, 0.2, 0.5, 1, 2, 5, 10)) # Instantiate
# ridge_model = RidgeCV(alphas=(0.01, 0.05, 0.1, 0.2, 0.5, 1, 2, 5, 10), normalize=True) # Instantiate
ridge_model.fit(X_train, y_train) #Fit
        
train_score=ridge_model.score(X_train,y_train)
test_score=ridge_model.score(X_test,y_test)
coeff_used = np.sum(ridge_model.coef_!=0)

print("best alpha: {}".format(ridge_model.alpha_))
print("training score: {}".format(train_score))
print("test score: {}".format(test_score))
print("number of features used: {}".format(coeff_used))
```

```python
coef = pd.Series(ridge_model.coef_, index = X_train.columns)
coef.sort_values(ascending=False)
```

```python
ridge_model = Ridge(alpha=0.1, normalize=True) # Instantiate
ridge_model.fit(X_train, y_train) #Fit
        
#Predict and score the model
y_train_preds = ridge_model.predict(X_train)
y_test_preds = ridge_model.predict(X_test)
    
train_score = r2_score(y_train, y_train_preds)
test_score = r2_score(y_test, y_test_preds)

#Print training and testing score
print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
```

```python
# use LassoCV to find best alpha automatically
from sklearn.linear_model import LassoCV

lasso_model = LassoCV(normalize=True, max_iter=10000) # Instantiate
lasso_model.fit(X_train, y_train) #Fit
        
train_score=lasso_model.score(X_train,y_train)
test_score=lasso_model.score(X_test,y_test)
coeff_used = np.sum(lasso_model.coef_!=0)

print("best alpha: {}".format(lasso_model.alpha_))
print("training score: {}".format(train_score))
print("test score: {}".format(test_score))
print("number of features used: {}".format(coeff_used))
```

```python
coef = pd.Series(lasso_model.coef_, index = X_train.columns)
coef.sort_values(ascending=False)
```
