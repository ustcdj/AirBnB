1. to explore if there is any seasonality of AirBNB home rental price
2. to compare rates between the two cities
3. to understand if there is anything about the properties that impact the price
4. Can you find negative and positive reviews based on text? 

```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option("display.max.columns", None)
pd.set_option('min_rows', 10)
# pd.options.display.max_rows
```

```python
df_boston_listings = pd.read_csv('boston\listings.csv', sep=',')
df_seattle_listings = pd.read_csv('seattle\listings.csv', sep=',')
```

```python
df_boston_listings.head(2)
```

```python
print(df_boston_listings.shape,  df_seattle_listings.shape)
```

```python
# the data structure from two cities are different. find the mismatch columns in two cities
df_boston_listings[df_boston_listings.columns.difference(df_seattle_listings.columns)].head()
```

```python
# concatenate data from two cities
df_listings = pd.concat([df_boston_listings, df_seattle_listings], keys=['Boston','Seattle'], names=['CITY','i'], sort=False)
df_listings.reset_index(inplace=True)
df_listings.set_index(keys=['id'], inplace=True, verify_integrity=True)
df_listings.drop(axis=1, inplace=True, columns=['i'])
```

```python
df_listings.head(2)
```

```python
# feat_info.csv was manually created and contains info about necessary property/actions to each column, such as id, 
# numeric, categorical, location, drop, convert to numeric and convert to date etc.
feat_info = pd.read_csv('feat_info.csv', sep=',')

# drop unnecessary columns
column_dropped_1 = feat_info[feat_info.action == 'drop']['name'].to_list()
df_listings.drop(columns = column_dropped_1, axis=1, inplace = True)
df_listings.rename(columns={'CITY': 'city'}, inplace=True)
df_listings.shape
```

```python
# obj to numeric
df_listings['price'] = df_listings['price'].replace('[$,]', '', regex=True).astype(float)
df_listings['host_acceptance_rate'] = df_listings['host_acceptance_rate'].str.replace('%', '').astype(float)/100
df_listings['host_response_rate'] = df_listings['host_response_rate'].str.replace('%', '').astype(float)/100

# convert to 0/1
df_listings['extra_people'] = df_listings['extra_people'].replace('[$,]', '', regex=True).astype(float)
df_listings['extra_people'] = df_listings['extra_people'].apply(lambda x: 1 if x>0 else 0)

# https://stackoverflow.com/questions/37543647/how-to-replace-all-non-nan-entries-of-a-dataframe-with-1-and-all-nan-with-0
df_listings[['security_deposit', 'cleaning_fee']] = df_listings[['security_deposit', 'cleaning_fee']].notnull().astype('int')

# calculate host days
df_listings['host_days'] = (pd.to_datetime(df_listings['last_scraped']) - pd.to_datetime(df_listings['host_since'])).dt.days

# convert calendar_updated to numeric days
calendar_updated_dict = {'today': 0, '2 weeks ago': 14, 'a week ago': 7, '3 weeks ago': 21, '2 months ago': 60, 'yesterday': 1, '4 weeks ago': 28, '3 months ago': 90, '4 days ago': 4, '5 weeks ago': 35, '3 days ago': 3, '5 days ago': 5, '4 months ago': 120, '6 weeks ago': 42, '2 days ago': 2, '5 months ago': 150, '7 weeks ago': 49, '7 months ago': 210, '6 days ago': 6, '1 week ago': 7, '6 months ago': 180, '8 months ago': 240, '10 months ago': 300, '9 months ago': 270, 'never': -1, '11 months ago': 330, '13 months ago': 390, '14 months ago': 420, '12 months ago': 360, '17 months ago': 510, '15 months ago': 450, '16 months ago': 480, '22 months ago': 660, '18 months ago': 540, '25 months ago': 750, '30 months ago': 900, '20 months ago': 600, '23 months ago': 690} 
df_listings.calendar_updated.replace(calendar_updated_dict, inplace=True)
df_listings['calendar_updated'] = np.where(df_listings['calendar_updated'] <0, df_listings['host_days'], df_listings['calendar_updated'])
```

```python
df_listings.hist(figsize=(16, 16))
```

## Discussion:
From the above histogram, it is easy to observe that calendar_updated, price, minimum_nights have a wide range of values. These values may obscure the final model. Thus, some rows with extreme values of these three variables will be deleted. 

```python
df_listings.price.hist(figsize=(15, 5), bins=60)
```

```python
# delete price outliers >=$800, 22 rows deleted
# df_listings.drop(df_listings[df_listings.price>800].index, axis=0, inplace=True)
df_listings = df_listings[df_listings.price<800]
df_listings.shape
```

```python
df_listings.calendar_updated.hist(figsize=(15, 5), bins=60)
```

```python
# delete calendar updated equal to or more than 365 days, 100 rows deleted
df_listings = df_listings[df_listings.calendar_updated<365]
df_listings.shape
```

```python
df_listings.minimum_nights.hist(figsize=(15, 5), bins=60)
```

```python
# delete minimum_nights equal to or more than 30 days, 49 rows deleted
# guests become tenants after 28 days
df_listings = df_listings[df_listings.minimum_nights<30]
df_listings.shape
```

```python
# missing data in each row
missing_row_ct = df_listings.isnull().sum(axis = 1)
missing_row_ct.sort_values(ascending = False, inplace=True)
missing_row_pct = missing_row_ct/(df_listings.shape[1]) * 100
# percentage of missing data in each row.
missing_row_pct.hist(figsize=(8, 4), bins=20)
plt.xlabel('% percentage of missing columns for each row')
```

```python
# delete rows with 30% more value missing, 2 rows deleted
df_listings = df_listings[df_listings.isnull().sum(axis=1) < df_listings.shape[1] * 0.3]
df_listings.shape
```

```python
# missing data in each column
missing_ct = df_listings.isnull().sum()
missing_ct.sort_values(ascending = False, inplace=True)
missing_pct = missing_ct/(df_listings.shape[0]) * 100
# percentage of missing data in each column.
missing_pct.plot.bar(figsize=(15, 4))
```

```python
df_listings.isnull().mean()[df_listings.isnull().mean() > 0].sort_values(ascending=False)
```

## Discussion
The majority of columns have no missing values. For those having missing values, no single column has missing value than 20%. Later, we will impute with mean or mode values. 

```python
df_listings[df_listings.isnull().mean()[df_listings.isnull().mean() > 0].index][df_listings.city=='Seattle'].hist(figsize=(15, 8))
```

```python
df_listings[df_listings.isnull().mean()[df_listings.isnull().mean() > 0].index][df_listings.city=='Boston'].hist(figsize=(15, 8))
```

```python
# fillna with mode for categorical columns, and the numeric columns that only take integer values
df_listings['host_response_time'].fillna(df_listings['host_response_time'].mode().iloc[0], inplace=True)
df_listings['property_type'].fillna(df_listings['property_type'].mode().iloc[0], inplace=True)
df_listings['bathrooms'].fillna(df_listings['bathrooms'].mode().iloc[0], inplace=True)
df_listings['bedrooms'].fillna(df_listings['bedrooms'].mode().iloc[0], inplace=True)
df_listings['beds'].fillna(df_listings['beds'].mode().iloc[0], inplace=True)
```

```python
df_listings.isnull().mean()[df_listings.isnull().mean() > 0].sort_values(ascending=False)
```

```python
df_listings[df_listings.city=='Seattle'].host_acceptance_rate.hist()
```

```python
df_listings[df_listings.city=='Boston'].host_acceptance_rate.hist()
```

```python
# fillna with mean for numeric columns
df_listings.fillna(df_listings.mean(), inplace=True)
```

```python
df_listings.drop(columns=['last_scraped','host_since'], axis=1, inplace=True)
```

```python
# convert 'amenities' to individual amenity with prefix amenities_
df_listings['amenities'].replace(to_replace='[\{\}\"]', value='', regex=True, inplace=True)
amenities_dummy = df_listings['amenities'].str.get_dummies(sep=',').add_prefix('amenities_')

# concatenate all individual amenities column with the full dataset
df_listings = pd.concat([df_listings, amenities_dummy], axis=1, sort=False)
df_listings.drop(columns=['amenities'], inplace=True, axis=1)
```

```python
cols_cat = df_listings.select_dtypes(include=['object']).columns.to_list()
```

```python
# One-hot encoding of categorical fetures
df_listings = pd.get_dummies(data=df_listings, prefix=cols_cat, prefix_sep='_', columns=cols_cat, drop_first=True)
```
